```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("startups.csv")
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
print(X)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
transformer = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder="passthrough")
X = np.array(transformer.fit_transform(X))
```

```python
print(X)
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(Xtrain, ytrain)
```

```python
y_pred = regressor.predict(Xtest)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), ytest.reshape(len(ytest),1)),1))
```

```python
for k in range(len(ytest)):
    y_teste = ytest[k]
    y_previsto = y_pred[k]
    erro = abs(((ytest[k] - y_pred[k])/ ytest[k]) *100)
    print(f'y_teste: {y_teste:.2f} y_previsto: {y_previsto:.2f} erro: {erro:.2f}')
```
