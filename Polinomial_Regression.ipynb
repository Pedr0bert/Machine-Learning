```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cargo_nivel_salarios.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df2 = df.drop('Cargo', axis=1)
```

```python
df2
```

```python
plt.scatter(df.Nivel, df.Salario, color='red')
plt.show()
```

```python
X = df2.iloc[:, :-1].values
y = df2.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
poly_regressor = PolynomialFeatures(degree=12)
```

```python
X_poly = poly_regressor.fit_transform(X)
```

```python
poly_regressor.fit(X_poly, y)
poly_regressor.fit(X,y)
lin_regressor = LinearRegression()
lin_regressor.fit(X_poly, y)
```

```python

plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, lin_regressor.predict(poly_regressor.fit_transform(X_train)), 'Db')
plt.show()
```

```python
plt.plot(X_test, y_test, "Dr")
plt.plot(X_test, lin_regressor.predict(poly_regressor.fit_transform(X_test)), color='blue')
plt.show()
```

```python
plt.scatter(X, y, color= "red")
plt.plot(X, lin_regressor.predict(poly_regressor.fit_transform(X)), "^b--")
plt.show()
```
